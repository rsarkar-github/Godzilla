## Parameters used for all the runs

```python
# Define frequency parameters (in Hertz)
freq_peak_ricker = 18
freq_max = 30
freq_min = 6.67
flat_spectrum = False
gaussian_spectrum = True
omega_max = 2 * Common.pi * freq_max
omega_min = 2 * Common.pi * freq_min
omega_mean = 2 * Common.pi * freq_peak_ricker
omega_std = (omega_max - omega_min) * 0.3
taper_pct = 0.1
dt = 0.5 / freq_max
nt = 100
domega = (2 * Common.pi) / (nt * dt)
delay = 0.1

# Create geometry object
geom2d = CreateGeometry2D(
    xdim=1.0,
    zdim=0.75,
    vmin=1.5,
    vmax=3.0,
    omega_max=omega_max,
    omega_min=omega_min
)
geom2d.set_default_params()
```

## Create the velocity plots

Create the velocity plots for background and background + reflectivity.

```python
# Create a default Velocity 2D object
vel_true = Velocity2D(geometry2d=geom2d)
vel_start = Velocity2D(geometry2d=geom2d)
ngridpoints_x = geom2d.gridpointsX
ngridpoints_z = geom2d.gridpointsZ

# Create starting velocity
vel_start.set_constant_velocity(vel=3.0)

# Put Gaussian perturbation in true model
sigma_x_gaussian = 0.1
sigma_z_gaussian = 0.1
center_nx = int(ngridpoints_x / 2)
center_nz = int(ngridpoints_z / 2.5)

vel_true.set_constant_velocity(vel=3.0)
vel_true.create_gaussian_perturbation(
    dvel=-1.5,
    sigma_x=sigma_x_gaussian,
    sigma_z=sigma_z_gaussian,
    nx=center_nx,
    nz=center_nz
)
vel = vel_true.vel
vel[:, center_nz + 69: center_nz + 70] = 2.0
vel_true.vel = vel
```

```python
%matplotlib inline

vel_true.plot(
    title="",
    pad=False,
    vmin=1.5,
    vmax=3.0,
    xlabel="X [km]",
    ylabel="Z [km]",
    show_colorbar=True,
    savefile="../Fig/velbkg-refl.pdf"
)

vel_start.plot(
    title="",
    pad=False,
    vmin=1.5,
    vmax=3.0,
    xlabel="X [km]",
    ylabel="Z [km]",
    show_colorbar=True,
    savefile="../Fig/velstart.pdf"
)
```

## Create the inverted images for 2 frequencies (20th and 40th, for 3 regularization values)

Create the  inverted images for 2 frequencies for the 3 regularization values.

```python
# Load files
with np.load('../Data/lsm-inverted-image-gacb-eps0.0.npz') as data:
    reg1_inversion = data['arr_0']

with np.load('../Data/lsm-inverted-image-gacb-eps0.1.npz') as data:
    reg2_inversion = data['arr_0']

with np.load('../Data/lsm-inverted-image-gacb-eps0.3.npz') as data:
    reg3_inversion = data['arr_0']

reg1_freq1 = np.real(reg1_inversion[19,:,:])
reg1_freq2 = np.real(reg1_inversion[39,:,:])

reg2_freq1 = np.real(reg2_inversion[19,:,:])
reg2_freq2 = np.real(reg2_inversion[39,:,:])

reg3_freq1 = np.real(reg3_inversion[19,:,:])
reg3_freq2 = np.real(reg3_inversion[39,:,:])
```

```python
import matplotlib.gridspec as gridspec

def plot_6_images(im1, im2, im3, im4, im5, im6, savefile):
    
    fig = plt.figure()
    fig.set_figheight(8)
    fig.set_figwidth(17)
    
    gs = gridspec.GridSpec(2, 3)
    gs.update(wspace=0.05, hspace=0.03)
    ax1 = plt.subplot(gs[0])
    ax2 = plt.subplot(gs[1])
    ax3 = plt.subplot(gs[2])
    ax4 = plt.subplot(gs[3])
    ax5 = plt.subplot(gs[4])
    ax6 = plt.subplot(gs[5])
    
    props = dict(boxstyle='square', facecolor='white', alpha=1.0)
    
    # Image 1
    vec = im1
    vmin = np.amin(vec)
    vmax = np.amax(vec)
    
    ax1.imshow(vec, origin="upper", vmin=vmin, vmax=vmax, cmap="Greys", interpolation="bilinear")
    ax1.set_ylabel("Z [km]", fontsize=14)
    
    xlim = ax1.get_xlim()
    ylim = ax1.get_ylim()

    ax1.set_xticklabels([])

    locs = ax1.get_yticks()
    labels = ["{:4.2f}".format(item * geom2d.dz) for item in locs]
    ax1.set_yticklabels(labels, fontsize=10)

    ax1.set_xlim(xlim)
    ax1.set_ylim(ylim)
    ax1.set_aspect("equal")
    
    ax1.text(0.05, 0.95, r'$\mathrm{\frac{\omega}{2 \pi} =}$' + " 15.54 Hz", 
             transform=ax1.transAxes, fontname='Arial', 
             fontsize=14, fontweight='normal', verticalalignment='top', bbox=props)
    
    # Image 2
    vec = im2
    vmin = np.amin(vec)
    vmax = np.amax(vec)
    
    ax2.imshow(vec, origin="upper", vmin=vmin, vmax=vmax, cmap="Greys", interpolation="bilinear")
    
    ax2.set_xticklabels([])
    ax2.set_yticklabels([])
    
    ax2.set_aspect("equal")
    
    ax2.text(0.05, 0.95, r'$\mathrm{\frac{\omega}{2 \pi} =}$' + " 15.54 Hz", 
             transform=ax2.transAxes, fontname='Arial', 
             fontsize=14, fontweight='normal', verticalalignment='top', bbox=props)
    
    # Image 3
    vec = im3
    vmin = np.amin(vec)
    vmax = np.amax(vec)
    
    ax3.imshow(vec, origin="upper", vmin=vmin, vmax=vmax, cmap="Greys", interpolation="bilinear")
    
    ax3.set_xticklabels([])
    ax3.set_yticklabels([])
    
    ax3.set_aspect("equal")
    
    ax3.text(0.05, 0.95, r'$\mathrm{\frac{\omega}{2 \pi} =}$' + " 15.54 Hz", 
             transform=ax3.transAxes, fontname='Arial', 
             fontsize=14, fontweight='normal', verticalalignment='top', bbox=props)
    
    # Image 4
    vec = im4
    vmin = np.amin(vec)
    vmax = np.amax(vec)
    
    ax4.imshow(vec, origin="upper", vmin=vmin, vmax=vmax, cmap="Greys", interpolation="bilinear")
    ax4.set_xlabel("X [km]", fontsize=14)
    ax4.set_ylabel("Z [km]", fontsize=14)
    
    xlim = ax4.get_xlim()
    ylim = ax4.get_ylim()

    locs = ax4.get_xticks()
    labels = ["{:4.2f}".format(item * geom2d.dx) for item in locs]
    ax4.set_xticklabels(labels, fontsize=10)

    locs = ax4.get_yticks()
    labels = ["{:4.2f}".format(item * geom2d.dz) for item in locs]
    ax4.set_yticklabels(labels, fontsize=10)

    ax4.set_xlim(xlim)
    ax4.set_ylim(ylim)
    ax4.set_aspect("equal")
    
    ax4.text(0.05, 0.95, r'$\mathrm{\frac{\omega}{2 \pi} =}$' + " 24.87 Hz", 
             transform=ax4.transAxes, fontname='Arial', 
             fontsize=14, fontweight='normal', verticalalignment='top', bbox=props)
    
    # Image 5
    vec = im5
    vmin = np.amin(vec)
    vmax = np.amax(vec)
    
    ax5.imshow(vec, origin="upper", vmin=vmin, vmax=vmax, cmap="Greys", interpolation="bilinear")
    ax5.set_xlabel("X [km]", fontsize=14)
    
    xlim = ax5.get_xlim()
    ylim = ax5.get_ylim()

    locs = ax5.get_xticks()
    labels = ["{:4.2f}".format(item * geom2d.dx) for item in locs]
    ax5.set_xticklabels(labels, fontsize=10)
    
    ax5.set_yticklabels([])

    ax5.set_xlim(xlim)
    ax5.set_ylim(ylim)
    ax5.set_aspect("equal")
    
    ax5.text(0.05, 0.95, r'$\mathrm{\frac{\omega}{2 \pi} =}$' + " 24.87 Hz", 
             transform=ax5.transAxes, fontname='Arial', 
             fontsize=14, fontweight='normal', verticalalignment='top', bbox=props)
    
    # Image 6
    vec = im6
    vmin = np.amin(vec)
    vmax = np.amax(vec)
    
    ax6.imshow(vec, origin="upper", vmin=vmin, vmax=vmax, cmap="Greys", interpolation="bilinear")
    ax6.set_xlabel("X [km]", fontsize=14)
    
    xlim = ax6.get_xlim()
    ylim = ax6.get_ylim()

    locs = ax6.get_xticks()
    labels = ["{:4.2f}".format(item * geom2d.dx) for item in locs]
    ax6.set_xticklabels(labels, fontsize=10)
    
    ax6.set_yticklabels([])

    ax6.set_xlim(xlim)
    ax6.set_ylim(ylim)
    ax6.set_aspect("equal")
    
    ax6.text(0.05, 0.95, r'$\mathrm{\frac{\omega}{2 \pi} =}$' + " 24.87 Hz", 
             transform=ax6.transAxes, fontname='Arial', 
             fontsize=14, fontweight='normal', verticalalignment='top', bbox=props)
    
    # Save images
    plt.savefig(savefile, bbox_inches="tight")
```

```python
plot_6_images(
    im1=reg1_freq1, 
    im2=reg2_freq1, 
    im3=reg3_freq1, 
    im4=reg1_freq2, 
    im5=reg2_freq2, 
    im6=reg3_freq2, 
    savefile="../Fig/inverted-images-gacb-comparison.pdf"
)
```
