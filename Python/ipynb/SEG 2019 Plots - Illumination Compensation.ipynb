## Parameters used for all the runs

```python
# Define frequency parameters (in Hertz)
freq_peak_ricker = 18
freq_max = 30
freq_min = 6.67
flat_spectrum = False
gaussian_spectrum = True
omega_max = 2 * Common.pi * freq_max
omega_min = 2 * Common.pi * freq_min
omega_mean = 2 * Common.pi * freq_peak_ricker
omega_std = (omega_max - omega_min) * 0.3
taper_pct = 0.1
dt = 0.5 / freq_max
nt = 100
domega = (2 * Common.pi) / (nt * dt)
delay = 0.1

# Create geometry object
geom2d = CreateGeometry2D(
    xdim=3.0,
    zdim=2.0,
    vmin=1.5,
    vmax=3.0,
    omega_max=omega_max,
    omega_min=omega_min
)
geom2d.set_default_params()
```

## Create the velocity plots

Create the velocity plots for noanomaly and with Gaussian anomaly.

### No Anomaly + Gaussian Anomaly - (background + reflectivity) models

```python
ngridpoints_x = geom2d.gridpointsX
ngridpoints_z = geom2d.gridpointsZ

# No anomaly
vel_true = Velocity2D(geometry2d=geom2d)

center_nz = int(ngridpoints_z / 2.5)

vel_true.set_constant_velocity(vel=3.0)
vel = vel_true.vel
vel[:, center_nz + 199: center_nz + 200] = 2.0
vel_true.vel = vel
vel_noanomaly = vel_true

# Gaussian anomaly
vel_true = Velocity2D(geometry2d=geom2d)

sigma_x_gaussian = 0.3
sigma_z_gaussian = 0.3
center_nx = int(ngridpoints_x / 2)
center_nz = int(ngridpoints_z / 2.5)

vel_true.set_constant_velocity(vel=3.0)
vel_true.create_gaussian_perturbation(
    dvel=-1.5,
    sigma_x=sigma_x_gaussian,
    sigma_z=sigma_z_gaussian,
    nx=center_nx,
    nz=center_nz
)
vel = vel_true.vel
vel[:, center_nz + 199: center_nz + 200] = 2.0
vel_true.vel = vel

vel_anomaly = vel_true
```

```python
%matplotlib inline

vel_noanomaly.plot(
    title="",
    pad=False,
    vmin=1.5,
    vmax=3.0,
    xlabel="X [km]",
    ylabel="Z [km]",
    show_colorbar=True,
    savefile="../Fig/veltrue-noanomaly.pdf"
)

vel_anomaly.plot(
    title="",
    pad=False,
    vmin=1.5,
    vmax=3.0,
    xlabel="X [km]",
    ylabel="Z [km]",
    show_colorbar=True,
    savefile="../Fig/veltrue-anomaly.pdf"
)
```

Create the inverted stacks with and without Gaussian anomaly.

```python
# Load files
with np.load('../Data/lsm-inverted-image-anomaly0.3-maxoff2.0-eps0.0_stack.npz') as data:
    anomaly_stack = data['arr_0']

with np.load('../Data/lsm-inverted-image-noanomaly-maxoff2.0-eps0.0_stack.npz') as data:
    noanomaly_stack = data['arr_0']
```

```python
def grey_image_plotting(vec, savefile):
    
    vmin = np.amin(vec)
    vmax = np.amax(vec)

    plt.figure()
    plt.imshow(vec, origin="upper", vmin=vmin, vmax=vmax, cmap="Greys", interpolation="bilinear")

    plt.xlabel("X [km]")
    plt.ylabel("Z [km]")
    plt.title("")

    xlim = plt.xlim()
    ylim = plt.ylim()

    locs, _ = plt.xticks()
    labels = ["{:4.2f}".format(item * geom2d.dx) for item in locs]
    plt.xticks(locs, labels)

    locs, _ = plt.yticks()
    labels = ["{:4.2f}".format(item * geom2d.dz) for item in locs]
    plt.yticks(locs, labels)

    plt.xlim(xlim)
    plt.ylim(ylim)

    plt.savefig(savefile, bbox_inches="tight")
```

```python
# Plot noanomaly stack
vec = np.real(noanomaly_stack)
grey_image_plotting(vec, "../Fig/noanomaly-inverted-image-stack.pdf")

# Plot anomaly stack
vec = np.real(anomaly_stack)
grey_image_plotting(vec, "../Fig/anomaly-inverted-image-stack.pdf")
```

```python
import matplotlib.gridspec as gridspec

def plot_4_images(im1, im2, im3, im4, savefile):
    
    fig = plt.figure()
    fig.set_figheight(10)
    fig.set_figwidth(15)
    
    gs = gridspec.GridSpec(2, 2)
    gs.update(wspace=0.05, hspace=0.05)
    ax1 = plt.subplot(gs[0])
    ax2 = plt.subplot(gs[1])
    ax3 = plt.subplot(gs[2])
    ax4 = plt.subplot(gs[3])
    
    props = dict(boxstyle='square', facecolor='white', alpha=1.0)
    
    # Image 1
    vec = im1
    vmin = np.amin(vec)
    vmax = np.amax(vec)
    
    ax1.imshow(vec, origin="upper", vmin=vmin, vmax=vmax, cmap="Greys", interpolation="bilinear")
    ax1.set_ylabel("Z [km]", fontsize=14)
    
    xlim = ax1.get_xlim()
    ylim = ax1.get_ylim()

    ax1.set_xticklabels([])

    locs = ax1.get_yticks()
    labels = ["{:4.2f}".format(item * geom2d.dz) for item in locs]
    ax1.set_yticklabels(labels, fontsize=10)

    ax1.set_xlim(xlim)
    ax1.set_ylim(ylim)
    ax1.set_aspect("equal")
    
    ax1.text(0.05, 0.95, r'$\mathrm{\frac{\omega}{2 \pi} =}$' + " 15.54 Hz", 
             transform=ax1.transAxes, fontname='Arial', 
             fontsize=14, fontweight='normal', verticalalignment='top', bbox=props)
    
    # Image 2
    vec = im2
    vmin = np.amin(vec)
    vmax = np.amax(vec)
    
    ax2.imshow(vec, origin="upper", vmin=vmin, vmax=vmax, cmap="Greys", interpolation="bilinear")
    
    ax2.set_xticklabels([])
    ax2.set_yticklabels([])
    
    ax2.set_aspect("equal")
    
    ax2.text(0.05, 0.95, r'$\mathrm{\frac{\omega}{2 \pi} =}$' + " 24.87 Hz", 
             transform=ax2.transAxes, fontname='Arial', 
             fontsize=14, fontweight='normal', verticalalignment='top', bbox=props)
    
    # Image 3
    vec = im3
    vmin = np.amin(vec)
    vmax = np.amax(vec)
    
    ax3.imshow(vec, origin="upper", vmin=vmin, vmax=vmax, cmap="Greys", interpolation="bilinear")
    ax3.set_xlabel("X [km]", fontsize=14)
    ax3.set_ylabel("Z [km]", fontsize=14)
    
    xlim = ax3.get_xlim()
    ylim = ax3.get_ylim()

    locs = ax3.get_xticks()
    labels = ["{:4.2f}".format(item * geom2d.dx) for item in locs]
    ax3.set_xticklabels(labels, fontsize=10)

    locs = ax3.get_yticks()
    labels = ["{:4.2f}".format(item * geom2d.dz) for item in locs]
    ax3.set_yticklabels(labels, fontsize=10)

    ax3.set_xlim(xlim)
    ax3.set_ylim(ylim)
    ax3.set_aspect("equal")
    
    ax3.text(0.05, 0.95, r'$\mathrm{\frac{\omega}{2 \pi} =}$' + " 15.54 Hz", 
             transform=ax3.transAxes, fontname='Arial', 
             fontsize=14, fontweight='normal', verticalalignment='top', bbox=props)
    
    # Image 4
    vec = im4
    vmin = np.amin(vec)
    vmax = np.amax(vec)
    
    ax4.imshow(vec, origin="upper", vmin=vmin, vmax=vmax, cmap="Greys", interpolation="bilinear")
    ax4.set_xlabel("X [km]", fontsize=14)
    
    xlim = ax4.get_xlim()
    ylim = ax4.get_ylim()

    locs = ax4.get_xticks()
    labels = ["{:4.2f}".format(item * geom2d.dx) for item in locs]
    ax4.set_xticklabels(labels, fontsize=10)
    
    ax4.set_yticklabels([])

    ax4.set_xlim(xlim)
    ax4.set_ylim(ylim)
    ax4.set_aspect("equal")
    
    ax4.text(0.05, 0.95, r'$\mathrm{\frac{\omega}{2 \pi} =}$' + " 24.87 Hz", 
             transform=ax4.transAxes, fontname='Arial', 
             fontsize=14, fontweight='normal', verticalalignment='top', bbox=props)
    
    # Save images
    plt.savefig(savefile, bbox_inches="tight")
```
