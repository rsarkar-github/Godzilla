```python
import numpy as np
import matplotlib.pyplot as plt
```

## Define a Gaussian
Let the Gaussian be $x(t) = e^{-at^2}$.
Its Fouier transform is $\hat{x}(\omega) = \int_{\mathbb{R}} x(t) e^{-i \omega t} dt = \sqrt{\frac{\pi}{a}}e^{- \frac{\omega^2}{4a}}$.

```python
# Define a Gaussian and time grids

a = 50                # controls width of Gaussian
T = 1                 # Domain is [-T, T]
N = 101               # Number of points (including endpoints)
dt = 2 * T / (N - 1)  # spacing in t

coords_t = np.linspace(-T, T, N, True)
x_t = np.exp(-1.0 * a* (coords_t ** 2))

plt.plot(coords_t, x_t, '-*b')
plt.grid(True)
plt.xlim(-T, T)
```

## Expected Fourier transform

```python
# Define nyquist frequency

f_nyq = 1.0 / (2 * dt)
omega_nyq = 2 * np.pi * f_nyq

# Calculate Fourier Transform on the grid [-omega_nyq : omega_nyq] with N samples
coords_omega = np.linspace(-omega_nyq, omega_nyq, N, True)
x_omega = ((np.pi / a) ** 0.5) * np.exp((-1.0 / (4 * a)) * (coords_omega ** 2) )

plt.plot(coords_omega, x_omega, '-*r')
plt.grid(True)
plt.xlim(-omega_nyq, omega_nyq)
```

## Fourier transform using DFT

```python
# Using DFT
x_omega_dft = dt * np.fft.fft(np.fft.fftshift(x_t))   # the factor of dt corrects for the spacing used in time

# Plot overlay
plt.plot(coords_omega, x_omega, '-*r')
plt.plot(coords_omega, np.fft.fftshift(x_omega_dft), '-*k')
plt.grid(True)
plt.xlim(-omega_nyq, omega_nyq)
```

```python
# Use of fftshift
u_t = np.linspace(-1,1,7,True)
aa = np.fft.fftshift(u_t)
bb = np.fft.ifftshift(aa)
plt.plot(aa,'-*b')
plt.plot(bb,'-*r')
plt.plot(u_t,'-*k')
```
